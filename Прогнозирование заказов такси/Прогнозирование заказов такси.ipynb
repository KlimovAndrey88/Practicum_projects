#### Построение графиков: исходного, тренда, сезонности и остатка за разные периоды времени (полгода, месяц и неделю). Построение графика сезонности за два дня.

**Построим графики за полгода**

```python
decomposition = seasonal_decompose(taxi_df['num_orders'], model='additive', period=24) #или 'multiplicative'

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

fig, axes = plt.subplots(4, 1, figsize=(15, 10))

axes[0].plot(taxi_df['num_orders'], label='Исходный временной ряд')
axes[0].legend()

axes[1].plot(trend, label='Тренд полугодовой')
axes[1].legend()

axes[2].plot(seasonal, label='Сезонность полугодовая')
axes[2].legend()

axes[3].plot(residual, label='Остаток')
axes[3].legend()

plt.tight_layout()
plt.show()
```

**Построим графики за месяц**

```python
decomposition_month = seasonal_decompose(taxi_df['num_orders'], model='additive', period=24) #или 'multiplicative'

trend_month = decomposition_month.trend['2018-07-31':'2018-08-31']
seasonal_month = decomposition_month.seasonal['2018-07-31':'2018-08-31']
residual_month = decomposition_month.resid['2018-07-31':'2018-08-31']

fig, axes = plt.subplots(4, 1, figsize=(15, 10))

axes[0].plot(taxi_df['num_orders'], label='Исходный временной ряд')
axes[0].legend()

axes[1].plot(trend_month, label='Тренд месячный')
axes[1].legend()

axes[2].plot(seasonal_month, label='Сезонность месячный')
axes[2].legend()

axes[3].plot(residual_month, label='Остаток')
axes[3].legend()

plt.tight_layout()
plt.show()
```

**Построим графики за неделю**

```python
decomposition_7 = seasonal_decompose(taxi_df['num_orders'], model='additive', period=24) #или 'multiplicative'

trend_7 = decomposition_7.trend['2018-08-24':'2018-08-31']
seasonal_7 = decomposition_7.seasonal['2018-08-24':'2018-08-31']
residual_7 = decomposition_7.resid['2018-08-24':'2018-08-31']

fig, axes = plt.subplots(4, 1, figsize=(15, 10))

axes[0].plot(taxi_df['num_orders'], label='Исходный временной ряд')
axes[0].legend()

axes[1].plot(trend_7, label='Тренд недельный')
axes[1].legend()

axes[2].plot(seasonal_7, label='Сезонность недельная')
axes[2].legend()

axes[3].plot(residual_7, label='Остаток')
axes[3].legend()

plt.tight_layout()
plt.show()
```

**Построим график сезонности за два дня**

```python
fig, ax = plt.subplots(figsize=(15, 10))

decomposed_day = seasonal_decompose(taxi_df)
ax.plot(decomposed_day.seasonal['2018-08-30':'2018-08-31'])

ax.set_title('Сезонность', fontsize=20)
ax.set_xlabel('Время', fontsize=15)
ax.set_ylabel('Значение', fontsize=15)
```

 1. Гистограмма имеет нормальное распределение со смещением влево и наличием пиковых значений. 
 2. По диаграмме размаха нормальный диапазон значений от 0 до 190. Среднее значение равно 84.4,  медиана примерно равна 80. Минимальное значение равно 0, максимальное - 462. Часть значений выходит за нормальный диапазон. 
 3. Среднее значение и стандартное отклонение увеличиваются (имеет положительный тренд).
 4. Тренды за все выбранные периоды времени имеют положительную тенденцию.
 5. Сезонность также прослеживается на всех графиках. Сезонность является суточной. График сезонности за два дня показывает зависимость от времени суток. С 00:00 до 06:00 количество заказов уменьшается, затем с 06:00 до 00:00 увеличивается с просадками в 12:00 и 18:00.
 
**Т.к. в датафрейме всего один количественных признак, нет необходимости в корреляционном анализе. Поэтому можно переходить к подготовке данных и обучению моделей.** 

## Обучение моделей

### Подготовка данных

#### Создадим новые признаки с помощью собственной функции

```python
def make_features(data, max_lag, rolling_mean_size):
    data = data.copy()
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    return data
```

```python
taxi_df_new = make_features(taxi_df, 24, 30)

taxi_df_new.head()
```

```python
taxi_df_new.shape
```

Новые признаки успешно созданы.

#### Разделим данные на тренировочную и тестовую выборки и входные и целевой признаки

```python
num_columns = taxi_df_new.select_dtypes(include=[np.number]).columns.tolist()
num_columns.remove('num_orders')

print("\nnum_columns:", num_columns)
```

```python
train, test = train_test_split(taxi_df_new, shuffle=False, test_size=0.1)

train = train.dropna()

# Вывод информации о train и test
print("\nTrain DataFrame:")
print(train.head())
print(train.info())

print("\nTest DataFrame:")
print(test.head())
print(test.info())

X_train = train.drop('num_orders', axis=1)
X_test = test.drop('num_orders', axis=1)

y_train = train['num_orders']
y_test = test['num_orders']

# Вывод столбцов X_train и X_test
print("\nX_train columns:", X_train.columns)
print("X_test columns:", X_test.columns)
```

#### Создадим пайплайн для подготовки данных

```python
#создаём пайплайн для подготовки признаков
num_pipe = Pipeline([
    ('imputer_num', SimpleImputer(missing_values=np.nan, strategy='median')),  # Или 'mean'
    ('scaler', StandardScaler())  # Или MinMaxScaler
])

#создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('num', num_pipe, num_columns)], 
    remainder='passthrough'
)
```

Данные подготовлены к обучению модели.

### Обучение и выбор лучшей модели

#### Создадим финальный пайплайн и param_grid для обучения моделей четырех классов: LinearRegression,  DecisionTreeRegressor, LGBMRegressor, CatBoostRegressor подбором гиперпараметров

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
])


param_grid = [{
    # словарь для модели LinearRegression()
        'models': [LinearRegression()],
        'preprocessor__num__imputer_num__strategy': ['median', 'mean'],
    },
    
    # словарь для модели DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=STATE)],
        'models__max_depth': range(1, 5),
        'models__min_samples_split': [16, 20],
        'models__min_samples_leaf': [12, 14],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },

    # словарь для модели LGBMRegressor()
    {
        'models': [LGBMRegressor(random_state=STATE)],
        'models__learning_rate': [0.1, 0.2],
        'models__num_leaves': [10, 20],
        'models__n_estimators': [20, 40],
        'models__reg_alpha': [0.01, 0.1, 1],
        'models__reg_lambda': [0.01, 0.1, 1],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },

    # словарь для модели CatBoostRegressor()
    {
        'models': [CatBoostRegressor(random_state=STATE, verbose=False)],
        'models__learning_rate': [0.1, 0.2],
        'models__depth': [1, 3],
        'models__l2_leaf_reg': [1, 5],
        'models__early_stopping_rounds': [20, 40],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'] 
    }
]    
```

```python
n_splits = 5 
ts_cv = TimeSeriesSplit(n_splits=n_splits)

randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid, 
    n_iter=25,
    cv=ts_cv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    error_score='raise'
)
randomized_search.fit(X_train, y_train)
```

Лучшая модель найдена с помощью RandomizedSearchCV и подбора гиперпараметров: модель класса CatBoostRegressor.

#### Выведем лучшую модель и метрику на кросс-валидации.

```python
print('Лучшая модель randomized_search и её параметры:\n\n', randomized_search.best_estimator_)
print('Метрика лучшей модели randomized_search на кросс-валидации:', randomized_search.best_score_)
```
